<h2>Filling the normalised losses is done through taking mean of categorical data of make</h2>

```python
df_final['normalised_losses'] = df_final.groupby(['make'])['normalised_losses'].transform(lambda x:x.fillna(x.mean()))
print df_final
```

<h2>Filling the price is done through taking mean of categorical data of make</h2>

```python
df_final['price'] =df_final.groupby(['make'])['price'].transform(lambda x:x.fillna(x.mean()))
print df_final
```
